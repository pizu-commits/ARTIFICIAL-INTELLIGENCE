## **MERGE AND CONTACT**

```python
var1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]}) # for related tables
var2=pd.DataFrame({"A":[1,2,3,4,6],"c":[11,21,31,41,51]})
pd.merge(var2,var1,on='A',how="outer",indicator=True)#left ,right,inner
```

```python
pd.merge(var2,var1,left_index=True,right_index=True,suffixes=("name","pyhotn"))
```

```python
s1=pd.Series([1,2,3,4,5,6]) # for same tables
s2=pd.Series([11,22,33,44,55])
pd.concat([s1,s2])
d1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})
d2=pd.DataFrame({"A":[1,2,3,4,6],"c":[11,21,31,41,51]})
pd.concat([d1,d2],axis=0,join="outer",keys=["d1",'d2'])#only inner and outer
```

## **JOIN and append(no error on same column name**

```python
d1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]})#cond:make index for both
d2=pd.DataFrame({"C":[1,2,3,4,6],"D":[11,21,31,41,51]})
d1.join(d2)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]},index=[1,2,3,4,5])#cond:make index for both
d2=pd.DataFrame({"C":[1,2,3,4,6],"D":[11,21,31,41,23]},index=[1,2,3,6,7])
d1.join(d2)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,22,33,44,55]},index=[1,2,3,4,5])#cond:make index for both
d2=pd.DataFrame({"C":[1,2,3,4,6],"B":[11,21,31,41,51]},index=[1,2,3,6,7])
d2.join(d1,how="right",lsuffix="1",rsuffix='2')
```
